```python
# Busquemos analizar que ocurre con la columna/variable/feature F_EX_DIR en funcion
# de la REGION

```

```python
# Analicemos unicamente a la region MOQUEGUA
moquegua = enedu2021_dir.loc[enedu2021_dir.REGION == "MOQUEGUA", : ]
moquegua_f_ex_dir = enedu2021_dir.loc[enedu2021_dir.REGION == "MOQUEGUA", "F_EX_DIR" ]
moquegua_f_ex_dir.plot(kind = "box")
```

```python
# Cuantos outliers (valores atipicos) tiene la region de MOQUEGUA  en la
# variable F_EX_DIR

moquegua = enedu2021_dir.loc[enedu2021_dir.REGION == "MOQUEGUA", : ]
Q1 = np.quantile(moquegua.F_EX_DIR, q = 0.25)
Q3 = np.quantile(moquegua.F_EX_DIR, q = 0.75)
IQR = Q3 - Q1

moquegua.loc[(moquegua.F_EX_DIR > Q3 + 1.5*IQR) | (moquegua.F_EX_DIR < Q1 - 1.5*IQR), : ].shape[0]

```

```python
  # Intervalo de valores para la variables F_EX_DIR por region
rango_F_EX_DIR_MOQUEGUA = moquegua.F_EX_DIR.max() -  moquegua.F_EX_DIR.min()
rango_F_EX_DIR_MOQUEGUA
```

```python
# Escalemos estas cuentas para todas las regiones
for region in enedu2021_dir.REGION.unique():
  # Separamos la informacion de una region en particular
  df_region = enedu2021_dir.loc[enedu2021_dir.REGION == region, : ]
  # Almacenamos en disco duro una imagen del boxplot para esta region
  # considerando como variables quantitativa : F_EX_DIR
  plt.figure()
  df_region.F_EX_DIR.plot(kind = "box")
  plt.title(region)
  # Calculo del numero de valores atipicos por region
  Q1 = np.quantile(df_region.F_EX_DIR, q = 0.25)
  Q3 = np.quantile(df_region.F_EX_DIR, q = 0.75)
  IQR = Q3 - Q1
  num_valores_atipicos_reg = df_region.loc[(df_region.F_EX_DIR > Q3 + 1.5*IQR) | (df_region.F_EX_DIR < Q1 - 1.5*IQR), : ].shape[0]
  # MOdifiquemos la etiqueta del eje-X agregando la informacion
  # del numero de valores atipicos
  plt.xlabel("Num. Outliers : " + str(num_valores_atipicos_reg))
  # Intervalo de valores para la variables F_EX_DIR por region
  rango_F_EX_DIR_region = df_region.F_EX_DIR.max() - df_region.F_EX_DIR.min()
  # Modifiquemos la etiqueta del eje-Y
  plt.ylabel("Rango : " + str(rango_F_EX_DIR_region))
  # Guardemos en disco duro esas imagenes
  # plt.savefig()


```
