# Estructura de decision : IF

```python
# if cond :
#   sentencia1
#   sentencia2
#   sentencia3
# elif cond2:
#   sentencia4
#   sentencia5
#   sentencia6
#   sentencia7
# elif cond3:
#   sentencia8
# else:
#   sentencia9
```

```python
# Verifiquemos la existencia de un triangulo de lados ingresados por el usuario
a = float(input("Ingresa el primer lado : "))
b = float(input("Ingresa el segundo lado : "))
c = float(input("Ingresa el tercer lado : "))

# Verifiquemos existencia :
if m.fabs(b-c) < a < (b+c) :
  if m.fabs(a-c) < b < (a+c) :
    if m.fabs(a-b) < c < (a+b) :
      print("""
      El triangulo de lados :
        a = %.2f
        b = %.2f
        c = %.2f
      EXISTE !!!
      """ %(a,b,c))
else:
  print("""
      El triangulo de lados :
        a = %.2f
        b = %.2f
        c = %.2f
      NO EXISTE !!!
      """ %(a,b,c))
```
